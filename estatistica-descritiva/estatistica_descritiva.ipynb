OBS2: 
   * count = contagem de linhas preenchidas
   * mean = média
   * std = desvio padrão (métrica de distribuição entorno da média)
   * min = limite inferior 
   * 25% = primeiro quartil
   * 50% = mediana
   * 75% = terceiro quartil
   * max = limite superior
  
  
OBS3: Sempre se atentar com os valores máximos e mínimos.

```python
# Entendendo os calculos por trás

l_numeros = [1,3,2,0,6]

media = sum(l_numeros) / len(l_numeros)
variancia = (((l_numeros[0]-media)**2) + ((l_numeros[1]-media)**2) + ((l_numeros[2]-media)**2) + ((l_numeros[3]-media)**2) + ((l_numeros[4]-media)**2)) / len(l_numeros)
desvio_padrao = variancia**(1/2)

print(media)
print(variancia)
print(desvio_padrao)
```

```python
# Calculando com numpy

import numpy as np

media = np.mean(l_numeros)
variancia = np.var(l_numeros)
desvio_padrao = np.std(l_numeros)

print(media)
print(variancia)
print(desvio_padrao)
```
